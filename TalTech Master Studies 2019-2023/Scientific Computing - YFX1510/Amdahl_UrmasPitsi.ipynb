This is a demonstration of Amdahl's Law.
Amdahl's law is often used in parallel computing to predict the theoretical speedup when using multiple processors. Depending on the hypothetical speedup factor p for a single task (in the range 0..1, where 1 means perfect speedup and no overhead), we can calculate the maximum speedup when executing the same task with n processors in parallel.
As we can see from the chart below, if speedup factor is even slightly below 1 (ie there is some overhead in combining parallel operations), the maximum achievable speedup converges to a constant value much lower than the number of processors doing the work. Conclusion: in practice there is a reasonable limit how many processors should be used in parallel. If the speedup factor of the single task is closer to 50% then maximum speedup is achived practically with 8 processors.

```python
import matplotlib.pyplot as plt
```

```python
def speedup(p=1, s=1):
    return 1 / (1 - p + p / s)
```

```python
num_procs_range = [i for i in range(1, 101)]
speedup_50 = [speedup(p=0.5, s=i) for i in num_procs_range]
speedup_75 = [speedup(p=0.75, s=i) for i in num_procs_range]
speedup_90 = [speedup(p=0.9, s=i) for i in num_procs_range]
speedup_95 = [speedup(p=0.95, s=i) for i in num_procs_range]
#speedup_99 = [speedup(p=0.9999, s=i) for i in num_procs_range]
```

```python
plt.plot(num_procs_range, speedup_50, label='50%')
plt.plot(num_procs_range, speedup_75, label='75%')
plt.plot(num_procs_range, speedup_90, label='90%')
plt.plot(num_procs_range, speedup_95, label='95%')
#plt.plot(num_procs_range, speedup_99, label='99%')
plt.title("Amdahl's Law")
plt.xlabel('Number of processors')
plt.ylabel('Speedup')
plt.legend()
```
